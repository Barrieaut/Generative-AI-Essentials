```python
!pip install transformers torch

import torch
from transformers import GPT2LMHeadModel, GPT2Tokenizer


# DATASET SELECTION & MODEL LOADING

# Dataset from Project Gutenberg (http://www.gutenberg.org/)
# load a pre-trained GPT-2 model which has been pre-trained on vast web text
model_name = "gpt2"
model = GPT2LMHeadModel.from_pretrained(model_name)
tokenizer = GPT2Tokenizer.from_pretrained(model_name)


# DEFINE A FUNCTION FOR TEXT GENERATION

def generate_text(prompt, max_length=100, temperature=1.0, top_k=50, top_p=0.95):
    input_ids = tokenizer.encode(prompt, return_tensors="pt")

    output = model.generate(
        input_ids,
        max_length=max_length,
        temperature=temperature,
        top_k=top_k,
        top_p=top_p,
        do_sample=True,
        num_return_sequences=1
    )

    generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
    return generated_text


# EXPERIMENT WITH DIFFERENT PROMPTS AND PARAMETERS


# Exercise 1: Basic text generation
prompt1 = "Once upon a time"
print("Exercise 1: Basic text generation")
print(generate_text(prompt1))
print("\n" + "="*50 + "\n")

# Exercise 2: Controlling output length
prompt2 = "The future of artificial intelligence is"
print("Exercise 2: Controlling output length")
print("Short Output (max_length=50):")
print(generate_text(prompt2, max_length=50))
print("\nLong Output (max_length=200):")
print(generate_text(prompt2, max_length=200))
print("\n" + "="*50 + "\n")

# Exercise 3: Adjusting temperature
prompt3 = "The benefits of renewable energy include"
print("Exercise 3: Adjusting temperature")
print("Low temperature (0.2):")
print(generate_text(prompt3, temperature=0.2))
print("\nHigh temperature (1.5):")
print(generate_text(prompt3, temperature=1.5))
print("\n" + "="*50 + "\n")

# Exercise 4: Using top_k and top_p
prompt4 = "In the year 2050, transportation will"
print("Exercise 4: Using top_k and top_p")
print("Default settings:")
print(generate_text(prompt4))
print("\nLow top_k and top_p:")
print(generate_text(prompt4, top_k=10, top_p=0.5))
print("\n" + "="*50 + "\n")

# Exercise 5: Prompt engineering
print("Exercise 5: Prompt engineering")
base_prompt = "Write a short story about"
topics = ["a robot learning to paint", "a time traveler's first day in the past", "a sentient AI's ethical dilemma"]

for topic in topics:
    full_prompt = f"{base_prompt} {topic}. The story should be creative and engaging:"
    print(f"Topic: {topic}")
    print(generate_text(full_prompt, max_length=150))
    print("\n" + "-"*30 + "\n")


# APPLICATION DEMONSTRATION: CONTENT CREATION

print("Application Demonstration: Content Creation")
content_prompt = "Product Description: A solar-powered smart jacket.\nMarketing Copy Generation:"
print(generate_text(content_prompt, max_length=120, temperature=0.7, top_k=40, top_p=0.9))
```

# Project Report: Fundamentals of Generative AI and GPT-2 Laboratory Analysis

## 1. Introduction to Generative AI
Generative Artificial Intelligence is a branch of computer science that creates new text, images, or media from patterns learned from training data. Unlike traditional machine learning systems that only classify data or predict static numbers, Generative AI generates realistic, novel sequences. This capability shifts computing from basic analysis tools to active human-centric content creation partners.

## 2. GPT Architecture and System Functionality
* **The Transformer Foundation**: Generative Pre-trained Transformers (GPT) utilize an architectural design that replaces sequential workflows with parallelized operations. This allows the system to analyze whole text strings simultaneously during processing phases.
* **Attention Mechanisms**: The self-attention matrix calculates operational weights between every token inside an input stream. This enables the model to resolve dependencies and retain full semantic context across long character sequences.
* **The Generation Loop**: Natural sentences are decoded into numerical arrays through tokenizers. The core model processes these index vectors to generate probability weights across its entire vocabulary table. Autoregressive sequence construction is maintained by appending newly selected tokens back into the active context loop.

## 3. Methodology and Parameter Insights
* **Dataset & Infrastructure**: Project Gutenberg (http://www.gutenberg.org/) text streams served as the open data baseline context. The project environment was initialized inside Google Colab using Python with PyTorch and the Hugging Face `transformers` ecosystem. We loaded a pre-trained base `gpt2` network infrastructure to execute runtime adjustments.
* **Length Constraints**: Modifying the `max_length` hyperparameter sets strict cutoffs for output sequences, preventing unbounded token execution.
* **Temperature Dynamics**: Low settings (e.g., 0.2) flatten vocabulary selection, making outputs highly repetitive and predictable. Elevated settings (e.g., 1.5) increase token variance, prompting creative but structurally unaligned language.
* **Top-K and Top-P Controls**: Top-K isolates selection to an exact number of highly ranked candidate tokens. Top-P dynamically caps candidate choices based on a cumulative percentage ceiling. Using both parameters simultaneously filters out rare or out-of-context words, keeping generated text realistic.

## 4. Practical Demonstration & Content Production Findings
* **Application Use Cases**: The script successfully implemented a real-world marketing content generation template to auto-create descriptions and commercial copy for a solar-powered product concept.
* **Operational Chokepoints**: Deploying generative transformers within customer-facing applications presents unique performance demands. Serving real-time text streams requires dedicated GPU instances to lower token latency, along with structured load balancing to handle spikes in request volumes.

## 5. Ethical Considerations & Operational Solutions
* **The Hallucination Dilemma**: Pre-trained language networks generate sequences based on structural probability rather than validated databases, which can lead them to invent incorrect information.
* **Data Bias Propagation**: Training models on open web text scrapes risks mirroring societal prejudices, toxicity, or discriminatory language patterns.
* **Mitigation Strategies**: Organizations can counter these issues by implementing human-in-the-loop validation, adding strict keyword filters to inputs and outputs, and anchoring model queries to verified internal databases using Retrieval-Augmented Generation (RAG).

## 6. Conclusion
This laboratory analysis validates how tuning specific execution parameters (such as temperature, top-k, and top-p) directly shapes the coherence and creativity of generated text. While base models like GPT-2 are prone to repetition and contextual drift over longer lengths, they establish the foundation for advanced modern text creation tools. Future iterations will focus on improving factual validation, lowering hosting resource overhead, and creating safer system alignment guidelines.
